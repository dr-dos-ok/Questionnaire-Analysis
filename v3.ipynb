```python
def calc_question (df, q):
    q = str(q)
    return (0* df['Count_Q'+q+'_A0'] + 1* df['Count_Q'+q+'_A1'] + 2* df['Count_Q'+q+'_A2'] + 3* df['Count_Q'+q+'_A3'] + 4*df['Count_Q'+q+'_A4'] +5*df['Count_Q'+q+'_A5']) / df['Count_Questionnaires']
```

```python
calc_question(y1425, 1)
```

```python
def calc_questionnaire (df):
    for i in range(29):
        return calc_question (df, i+1)

# calc_questionnaire(y1425)
```
